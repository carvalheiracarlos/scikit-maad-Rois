#### Imports

```python
import gdown
import pandas as pd
import tensorflow as tf
from maad import sound, rois, features
```

#### Constants

```python
drive_url = 'https://drive.google.com/drive/folders/1qUo8aAwrpRVtYrdJHu3j--OnyjUtGu7N'
audios = './audios'
files_txt = './audios/PMpiloto_20181121_214822-220540_assoviosTursiops.txt'
sampling_rate = 16000
```

#### Download from Google Drive

```python
def download_data_from_google():
  gdown.download_folder(url=drive_url, 
                        output=audios,
                        quiet=False, 
                        use_cookies=False
                        )
```

```python
download_data_from_google()
```

#### Load Audio from files

```python
metadata = pd.read_csv(files_txt, sep='\t')
metadata[61:]
```

```python
def load_audio(file_paths):
  for audio_file in file_paths:
    record_name  = audio_file.split('/')[2]
    record_stats = metadata.loc[metadata['Begin File'] == record_name]
    print(f'Processing File: {record_name}')
    for index, row in record_stats.iterrows():
      audiogram, sampling_fq = sound.load(audio_file)
      rois.find_rois_cwt(audiogram, 
                         sampling_fq, 
                         flims=(row['Low Freq (Hz)'],
                                row['High Freq (Hz)']
                              ), 
                         tlen=row['End Time (s)'] - row['Begin Time (s)'], 
                         th=0,
                         display=False,
                         save_df=True
                        ) 
```

```python
listed_files = audios + '/' + metadata['Begin File']
```

```python
load_audio(listed_files)
```
